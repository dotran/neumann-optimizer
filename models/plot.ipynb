```python
import _pickle as pkl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('./32_5.pkl', "rb") as f:
    losses_32 = pkl.load(f)
```

```python
with open('./256_5.pkl', "rb") as f:
    losses_256 = pkl.load(f)
```

```python
with open('./2048_5.pkl', "rb") as f:
    losses_2048 = pkl.load(f)
```

```python
plt.figure(figsize=(12, 8))
plt.title("Neumann Opt on Action Detection", fontsize=17)
plt.xlabel("Iteration", fontsize=15)
plt.ylabel("Loss", fontsize=15)
plt.xlim(0, 58)
plt.ylim(1.6, 2.4)
plt.plot(losses_32, label="Batch size 32")
plt.plot(losses_256, label="Batch size 256")
plt.plot(losses_2048, label="Batch size 2048")
plt.legend(fontsize=15)
```
